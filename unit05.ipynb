```python
import csv
import matplotlib.pyplot as plt

HOT = -1
result = []
cnt = 0

f = open('seoul.csv')
data = csv.reader(f)

header = next(data)

for row in data:
    cnt = cnt + 1
    if row[HOT] != '':
        result.append(float(row[HOT]))
plt.figure(figsize=(10,2), dpi=300)
plt.plot(result, 'r')
plt.show()

print(cnt)
```

```python
import csv
import matplotlib.pyplot as plt

HOT = -1
COLD = -2
DATE = 0
YEAR = 0
MON = 1
DAY = 2

high = []
low = []
years = []
cnt = 0

f = open('seoul.csv')
data = csv.reader(f)

header = next(data)

for row in data:
    cnt = cnt + 1
    if row[HOT] != '':
        if int(row[DATE].split('-')[YEAR]) >= 1983 and row[DATE].split('-')[MON] == '01' and row[DATE].split('-')[DAY] == '08':
            years.append(int(row[DATE].split('-')[YEAR]))
            high.append(float(row[HOT]))
            low.append(float(row[COLD]))
plt.figure(figsize=(10,2), dpi=300)
plt.title("      ")
plt.plot(years, high, 'hotpink', label="High")
plt.plot(years, low, 'skyblue', label="Low")
plt.rc('font', family='Malgun Gothic')      #   
plt.rcParams['axes.unicode_minus'] = False #    
plt.legend()
plt.show()
```
