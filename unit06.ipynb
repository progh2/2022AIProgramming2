```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            result.append(float(row[-1]))
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.hist(result, bins=100, color='r')
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        elif month == '08':
            aug.append(float(row[-1]))
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = int(row[0].split('-')[1])
    if row[-1] != '':
        if month == 1:
            jan.append(float(row[-1]))
        elif month == 8:
            aug.append(float(row[-1]))
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.style.use('ggplot')
plt.boxplot([aug,jan])
plt.legend()
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
months = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    month = int(row[0].split('-')[1])
    if row[-1] != '':
        months[month - 1].append(float(row[-1]))
        
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.boxplot(months)
plt.legend()
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
months = [[],[],[],[],[],[],[],[],[],[],[],[]]
days = []
for i in range(31):
    days.append([])

for row in data:
    month = int(row[0].split('-')[1])
    day = int(row[0].split('-')[2])
    if row[-1] != '':
        if month == 8:
            days[day - 1].append(float(row[-1]))
        
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.boxplot(days, showfliers=False)
plt.legend()
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
months = [[],[],[],[],[],[],[],[],[],[],[],[]]
days = []
for i in range(31):
    days.append([])

for row in data:
    month = int(row[0].split('-')[1])
    day = int(row[0].split('-')[2])
    if row[-1] != '':
        if month == 8:
            days[day - 1].append(float(row[-1]))
        
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.boxplot(days, showfliers=False)
plt.show()
```
