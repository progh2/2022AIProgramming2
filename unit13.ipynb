```python
import matplotlib.pyplot as plt
t = []
p2 = []
p3 = []
for i in range(0, 50, 2):
    t.append(i/10)
    p2.append((i/10)**2)
    p3.append((i/10)**3)

plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.rc('font', family='Malgun Gothic')
plt.title('리스트를 활용한 코드')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.,5.,0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.rc('font', family='Malgun Gothic')
plt.title('리스트를 활용한 코드')
plt.show()
```

```python
print(np.random.choice(6,10))  # 0부터 6미만 정수 10개 생성. 콤마가 없음. ndarray타입이다!!
```

```python
import matplotlib.pyplot as plt
import numpy as np
dice = np.random.choice(6,10)
plt.figure(dpi=300)
plt.hist(dice, bins=6)
plt.show()
```
