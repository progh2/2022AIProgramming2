```python
import csv

f = open('gender3.csv')
data = csv.reader(f)
headers = next(data)
M_INIT = -1
F_INIT = -1
m = []
f = []

i = 0
for header in headers:
    #print(i, ':', header)
    if "남_0세" in header:
        M_INIT = i
    elif "여_0세" in header:
        F_INIT = i
    i+= 1
    if M_INIT != -1 and F_INIT != -1:
        break;
        
# print(M_INIT, F_INIT)

name = input('인구 구조가 알고 싶은 지역의 이름을 입력해 주세요(생략 시 강내면 검색): ')
if name == "":
    name = "강내면"

for row in data:
    if name in row[0]:
        #print(row)
        for i in row[ M_INIT:(M_INIT+100)]:
            m.append(-1 * int(i.replace(",","")))
        for i in row[ F_INIT:(F_INIT+100)]:
            f.append(int(i.replace(",","")))
        break
#print(m)
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5), dpi=300)
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
plt.title(name + " 지역의 남녀 성별 인구 분포")
#plt.plot(m)
plt.barh(range(len(m)), m, label='남성')
plt.barh(range(len(f)), f, label='여성')
plt.legend()
plt.show()
```
