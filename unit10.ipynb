```python
import csv
import math

file = open('gender3.csv')
data = csv.reader(file)
headers = next(data)
M_INIT = -1
F_INIT = -1
m = []
f = []
size = []
m_sum = 0
f_sum = 0

i = 0
for header in headers:
    #print(i, ':', header)
    if "남_0세" in header:
        M_INIT = i
    elif "여_0세" in header:
        F_INIT = i
    i+= 1
    if M_INIT != -1 and F_INIT != -1:
        break;
        
# print(M_INIT, F_INIT)
name = "강내면"
isLoop = True
while(isLoop):
    name = ""
    name = input('인구 구조가 알고 싶은 지역의 이름을 입력해 주세요(생략 시 강내면 검색): ')
    if name == "":
        name = "강내면"

    # 중복 검색 지역이 있는지 확인. 만약 있다면 다시 입력받음
    name_count = 0
    name_list = []
    for row in data:
        if name in row[0]:
            name_count += 1
            name_list.append(row[0])

    file.seek(1)
    if name_count == 1:
        isLoop = False
    elif name_count == 0:
        print("해당 검색명을 가진 지역이 없습니다. 올바른 이름으로 검색해주세요!!!")
    else:
        print("다음과 같이 여러 지역이 검색되었습니다. 좀 더 구체적으로 입력해주세요!!")
        for item in name_list:
            print(item)

for row in data:
    if name in row[0]:
        #print(row)
        for i in row[ M_INIT:(M_INIT+100)]:
            m.append(int(i.replace(",","")))
            m_sum += int(i.replace(",",""))

        for i in row[ F_INIT:(F_INIT+100)]:
            f.append(int(i.replace(",","")))
            f_sum += int(i.replace(",",""))
            
        for i in range(100):
            size.append(math.sqrt( m[i] + f[i] ) )
        break
#print(m)
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5), dpi=300)
plt.style.use('ggplot')
plt.scatter(m,f)
plt.scatter(m,f, s=size, c=range(len(m)), cmap='jet', alpha=0.5 )
plt.rcParams['axes.grid'] = False
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.xlabel('남성 인구 수')
plt.ylabel('여성 인구 수')
plt.title(name + " 지역의 성별 인구 그래프")
plt.show()
```
