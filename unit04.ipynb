```python
plt.plot([10, 20, 30, 40])
plt.title('plotting')
plt.show()
```

```python
# 범례 넣기. 라벨로 지정
plt.title('legend')
plt.plot([10,20,30,40], label='asc') # asc 범뢰
plt.plot([40,30,20,10], label='desc') # asc 범뢰
plt.legend()
plt.show()
```

```python
# 범례 넣기. 라벨로 지정
plt.title('color')
plt.plot([10,20,30,40], label='skyblue', color='skyblue') # asc 범뢰
plt.plot([40,30,20,10], 'pink', label='pink') # asc 범뢰
plt.legend()
plt.show()
```

```python
plt.title('marker')
plt.plot([10,20,30,40], 'r.', label='circle') # 빨간색 원형 마커 그래프
plt.plot([40,30,20,10], 'g^', label='triangle up')
plt.legend() # 범례 표시
plt.show()
```

```python
plt.title('marker')
plt.plot([10,20,30,40], 'r.--', label='dashed') # 빨간색 원형 마커 그래프
plt.plot([40,30,20,10], 'g^:', label='dotted')
plt.legend()
plt.show()
```
