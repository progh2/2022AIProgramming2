```python
plt.plot([10, 20, 30, 40])
plt.title('plotting')
plt.show()
```

```python
#  .  
plt.title('legend')
plt.plot([10,20,30,40], label='asc') # asc 
plt.plot([40,30,20,10], label='desc') # asc 
plt.legend()
plt.show()
```

```python
#  .  
plt.title('color')
plt.plot([10,20,30,40], label='skyblue', color='skyblue') # asc 
plt.plot([40,30,20,10], 'pink', label='pink') # asc 
plt.legend()
plt.show()
```

```python
plt.title('marker')
plt.plot([10,20,30,40], 'r.', label='circle') #    
plt.plot([40,30,20,10], 'g^', label='triangle up')
plt.legend() #  
plt.show()
```

```python
plt.title('marker')
plt.plot([10,20,30,40], 'r.--', label='dashed') #    
plt.plot([40,30,20,10], 'g^:', label='dotted')
plt.legend()
plt.show()
```
