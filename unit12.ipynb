```python
import csv
f = open('subwaytime.csv')
data = csv.reader(f)
header1 = next(data)
header2 = next(data)

NUM = 24
mx = [0] * NUM
mx_station = [''] * NUM

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        in_cnt = row[j * 2 + 4]
        if mx[j] < in_cnt:
            mx[j] = in_cnt
            time = j + 4
            if time > 24:
                time -= 24
            mx_station[j] = row[1] + row[3] + '(' + str(time) + '시)'

for i in range(NUM):
    time = i + 4
    if time > 24:
        time -= 24
    print( time,"시 : ", mx_station[i], mx[i])
        
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.figure(dpi=300)
plt.bar(range(NUM), mx)
plt.xticks(range(NUM), mx_station, rotation = 90)
plt.show()
```

```python
import csv
f = open('subwaytime.csv')
data = csv.reader(f)
header1 = next(data)
header2 = next(data)

NUM = 24
s_in = [0] * NUM
s_out = [0] * NUM

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(NUM):
        s_in[j] += row[j * 2 + 4]
        s_out[j] += row[j * 2 + 5]

for i in range(NUM):
    time = i + 4
    if time > 24:
        time -= 24
    print( time,"시 : 승차", s_in[i], "하차", s_out[j])

times = [''] * NUM
for i in range(NUM):
    times[i] = i + 4
    if (times[i] > 24):
        times[i] -= 24
    times[i] = str(times[i]) + "시"
        
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label="승차", color="skyblue")
plt.plot(s_out, label="하차", color="pink")
plt.legend()
plt.xticks(range(NUM), times, rotation = 90)
plt.show()
```
